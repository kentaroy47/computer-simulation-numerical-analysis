# 例題4-1

$f(x) = x^2 - 2 = 0$の解を二分法で求める

```python
import numpy as np

a = np.float32(5.)
b = np.float32(0)

def f(x):
    return np.power(x, 2) - 2

for i in range(20):
    c = (a+b)/2
    if f(c) < 0:
        b = c
    else:
        a = c
    print(i+1, np.sqrt(2)-c)
```

# 4.2

上記の二分法について中間値と誤差の関係をグラフにしなさい

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.float32(5.)
b = np.float32(0)

# 結果格納用のlistを用意
iters = []
cs = []
errors = []

def f(x):
    return np.power(x, 2) - 2

for i in range(20):
    c = (a+b)/2
    if f(c) < 0:
        b = c
    else:
        a = c
    print(i+1, np.sqrt(2)-c)
    iters.append(i+1)
    errors.append(np.sqrt(2)-c)

# グラフにプロット
plt.plot(iters, errors)
plt.xlabel("試行回数", fontname = 'MS Gothic')#日本語用
plt.ylabel("誤差", fontname = 'MS Gothic')
```

```python
# グラフにプロット
plt.plot(iters[15:], errors[15:])
plt.xlabel("試行回数", fontname = 'MS Gothic')#日本語用
plt.ylabel("誤差", fontname = 'MS Gothic')
```
