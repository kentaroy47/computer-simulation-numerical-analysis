# 課題10-1

```python
import numpy as np
import matplotlib.pyplot as plt
maxt = np.pi*4

u = 0; y = 1; h = 0.25
ys = []; us = []; yas = []
for i in np.arange(0, maxt, h):
    ya = np.cos(i)
    temp = u
    u = -y * h + u
    y = temp * h + y
    us.append(u); yas.append(ya); ys.append(y)
plt.plot(np.arange(0, maxt, 0.25), ys, label='y')
plt.plot(np.arange(0, maxt, 0.25), yas, label='analytical')
plt.legend()    
```

# 課題10-3

```python
u = 0; y = 1; h = 0.25
print(h)
ys = []; us = []; yas = []

for i in np.arange(h, maxt, h):
    ya = np.cos(i)
    k1 = -y; m1 = u
    u1 = k1*h/2 + u
    y1 = m1*h/2 + y
    k2 = -y1; m2 = u1
    u2 = k2*h/2 + u
    y2 = m2*h/2 + y
    k3 = -y2; m3 = u2
    u3 = k3*h + u
    y3 = m3*h + y
    k4 = -y3; m4 = u3
    u = (k1+2*k2+2*k3+k4)/6*h + u
    y = (m1+2*m2+2*m3+m4)/6*h + y
    yas.append(ya); ys.append(y)
plt.plot(np.arange(h, maxt, h), ys, label='y')
plt.plot(np.arange(h, maxt, h), yas, label='analytical')
plt.legend()    
```
