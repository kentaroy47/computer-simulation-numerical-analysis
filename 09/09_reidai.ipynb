# 例題9-1

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x

y = 1
ye = y
yh = y
yr = y
yes = []; yhs = []; yrs = []
dt = 0.2
for i in np.arange(0, 5, 0.2):
    # オイラー法
    ye = ye*dt + ye
    yes.append(ye)
    # ホイン法
    k1 = f(yh)
    yh1 = k1*dt + yh
    k2 = f(yh1)
    yh = (k1+k2)*dt/2 + yh
    yhs.append(yh)
    # ルンゲクッタ法
    k1 = f(yr)
    yr1 = k1*dt/2 + yr
    k2 = f(yr1)
    yr2 = k2*dt/2 + yr
    k3 = f(yr2)
    yr3 = k3*dt + yr
    k4 = f(yr3)
    yr = (k1+2*k2+2*k3+k4)*dt/6 + yr
    yrs.append(yr)
plt.plot(np.arange(0, 5, 0.2), yes, label="Euler")
plt.plot(np.arange(0, 5, 0.2), yhs, label="Heun")
plt.plot(np.arange(0, 5, 0.2), yrs, label="Runge-Kutta")
plt.legend()
```

# 8.2
$sinx$の$x=1$における二階微分係数を求めなさい

```python
x = 1
df0 = -np.sin(x) # 解析値
for i in range(6):
    h = 10**(-i)
    # 前進差分
    df1 = (f(x+2*h) - 2*f(x+h) + f(x))/np.power(h, 2)
    # 後退差分
    df2 = (f(x) - 2*f(x-h) + f(x-2*h))/np.power(h, 2)
    # 中心差分
    df3 = (f(x+h) - 2*f(x) + f(x-h))/np.power(h, 2)
    print("h=",h," 前進差分誤差：", abs(df1-df0), " 後退差分誤差：", abs(df2-df0), " 中心差分誤差：", abs(df3-df0))
```

# 7.3
シンプソン法によって積分値を求め、不定積分の値と比較せよ

```python
a = 0; b = 2; N = 16

h = (b - a) / N
sum = 0.0

for i in range(0, N, 2):
    x1 = a + i * h
    x2 = a + (i + 1) * h
    x3 = a + (i + 2) * h
    sum += (h / 3) * (f(x1) + 4 * f(x2) + f(x3))

exact_integral = np.sin(2)
numerical_integral = sum
difference = abs(exact_integral - numerical_integral)

print("Exact integral value: ", exact_integral)
print("Numerical integral value: ", numerical_integral)
print("Difference: ", difference)
```
